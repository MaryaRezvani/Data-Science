```python
# Perform the Shapiro-Wilk test for normality.
from scipy.stats import shapiro
import numpy as np
```

```python
data = np.random.normal(loc=20, scale=5, size=1000)
print(data)
```

```python
# If the data falls below the 5% threshold(pvalue < 0.5), it does not follow a normal distribution.

shapiro(data)
```

```python
import scipy.stats as stats
import pylab
```

```python
# Generates a probability plot of sample data against the quantiles of a specified theoretical distribution (the normal distribution by default)

stats.probplot(data, dist='norm', plot=pylab)
# _,fit=stats.probplot(data, dist=stats.norm, plot=pylab)
# goodness_fit="%.2f" %fit[2]
```

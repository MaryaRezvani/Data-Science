```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker
import os
```

```python
src_file = os.path.join('EPA_fuel_economy.csv')
image_dir = ('../images')
```

```python
df = pd.read_csv(src_file)
df.head()
```

```python
df.info()
```

```python
plt.hist(df['Combined MPG (FT1)'])
```

```python
plt.hist(df['Combined MPG (FT1)'])
plt.xlabel('Combined Fuel Economy(Mile Per Gallon)')
plt.ylabel(('Number of cars'))
plt.title('Combined Fuel Economy')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df['Combined MPG (FT1)'])
```

```python
fig, ax = plt.subplots()
ax.hist(df['Combined MPG (FT1)'])
ax.set_xlabel('Combined Fuel Economy(Mile Per Gallon)')
ax.set_ylabel('Number of cars')
ax.set_title('Combined Fuel Economy')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df['Combined MPG (FT1)'])
ax.set(xlabel ='Combined Fuel Economy(Mile Per Gallon', ylabel = 'Number of cars',
title = 'Combined Fuel Economy')
```

```python
fig, ax = plt.subplots()
ax.hist(df['Combined MPG (FT1)'], range(10, 50),histtype='step', orientation='horizontal')
```

```python
fig, ax = plt.subplots()
ax.boxplot(df['Combined MPG (FT1)'], labels=['All'])
ax.set(title='MPG Distribution', ylabel='MPG for Vehicle')
```

```python
fig, ax = plt.subplots()
ax.boxplot(df['Combined MPG (FT1)'], labels=['All'], showfliers=False)
ax.set(title='MPG Distribution', ylabel='MPG for Vehicle')
```

```python
fig, (ax1, ax2) = plt.subplots(2)
ax1.hist(df['City MPG (FT1)'], range(10,50))
ax1.set(title='Vehicle MPG', xlabel='Combined MPG', ylabel='Number of Cars')

ax2.boxplot(df['Combined MPG (FT1)'], labels=['All'], vert=False, showfliers=False)
ax2.set(ylabel='MPG for vehicles');
fig.suptitle('MPG Distribution', fontsize=14, fontweight='bold')
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(9,4))
ax1.hist(df['City MPG (FT1)'], range(10,50))
ax1.set(title='Vehicle MPG', xlabel='Combined MPG', ylabel='Number of Cars')

ax2.boxplot(df['Combined MPG (FT1)'], labels=['All'], showfliers=False)
ax2.set(ylabel='MPG for vehicles');
fig.suptitle('MPG Distribution', fontsize=14, fontweight='bold')
```

```python
fig.savefig(os.path.join(image_dir, 'mpg.png'), transparent=True, dpi=80, bbox_inches="tight")
```

```python

avg_by_year = df.groupby(['Year'], as_index=False).agg({'Combined MPG (FT1)':'mean'}).round(2)
avg_by_year
```

```python
fig, ax1 = plt.subplots()
ax1.plot(avg_by_year['Year'], avg_by_year['Combined MPG (FT1)']);
```

```python


fig, ax1 = plt.subplots()
ax1.plot(avg_by_year['Year'], avg_by_year['Combined MPG (FT1)']);
ax1.set_xticks(np.arange(1985, 2020, 3));
```

```python
fig, ax1 = plt.subplots()
ax1.bar(avg_by_year['Year'], avg_by_year['Combined MPG (FT1)']);
ax1.set_xticks(np.arange(1985, 2020, 3));
```

```python
fig, ax1 = plt.subplots()
ax1.barh(avg_by_year['Year'], avg_by_year['Combined MPG (FT1)']);
```

```python


fig, ax1 = plt.subplots()
ax1.scatter(x=df['Annual Fuel Cost (FT1)'], y=df['Engine Displacement'], alpha=0.5, c=df['Engine Cylinders']);
ax1.set(xlabel='Cost', ylabel='Displacement')
```

```python


fig, ax1 = plt.subplots(figsize=(9,7))
ax1.scatter(x=df['Annual Fuel Cost (FT1)'], y=df['Engine Displacement'], alpha=0.5, c=df['Engine Cylinders']);
ax1.set_xlabel('Cost', size=14)
ax1.set_ylabel('Displacement', size=14)
ax1.xaxis.set_major_formatter('${x:,.0f}')
ax1.tick_params(axis='x', labelrotation=45, labelsize=14)
ax1.tick_params(axis='y', labelsize=14)
ax1.axvline(3500, color='black', linestyle=':')
ax1.annotate('Target of $3500', xy=(3500, 2), size=16)
ax1.grid(True)
```
